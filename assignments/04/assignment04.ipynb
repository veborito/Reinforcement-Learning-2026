## 1.1

Super simple MDP with 10 states. Agent move in a straight line. Deciedes if he wants
to move or stay at the same place. If he stays it's a 0 cost and if he moves it costs him 1 (-1 reward). Arriving half way he gets 10 reward and if he
manages to go to the end state he will receive a huge reward.

Probability that the agent gets to the next state if he decides to move is 0.7
And with 0.3 probability he will go back to starting position.

## 1.2 Implementation

DiscreteMDP, ValueIteration, Policy Iteration code were taken from the rldmuu github.

```python
from MDP import DiscreteMDP


class MySuperSimpleMDP(DiscreteMDP):
  def __init__(self, n_states):
    
    super().__init__(n_states=n_states, n_actions=2)
    
    self.STAY = 0
    self.MOVE = 1
    
    self.P[:] = 0
    self.R[:] = 0
    self.R[:, self.MOVE] = -1
    self.R[:, self.STAY] = 0
    
    self.R[n_states - 1,:] = 100
    self.R[n_states // 2 - 1,:] = 10
    
    for s in range(n_states):
      self.P[s, self.STAY, s] = 1
      
      if s == n_states - 1:
        self.P[s, self.MOVE, s] = 1
      else:
        self.P[s, self.MOVE, s + 1] = 0.7
        self.P[s, self.MOVE, 0] = 0.3
```

## 1.3

```python
mdp = MySuperSimpleMDP(10)
```

```python
from ValueIteration import value_iteration
import numpy as np

for gamma in np.arange(0.1, 1, 0.1):
  policy, V, Q = value_iteration(mdp, 1000, gamma)
  print(f"gamma: {gamma} policy:")
  print(policy)
  print("Value function")
  print(V)

```

```python
from PolicyIteration import policy_iteration

for gamma in np.arange(0.1, 1, 0.1):
  policy, V, Q = policy_iteration(mdp, 10000, gamma, 100)
  print(f"gamma: {gamma} policy:")
  print(policy)
```

### Discussion

With this super simple MDP example, we clearly see that the policy depends on gamma.\
In the value iteration algorithm.\
Same conclusions with the policy iteration algorithm

In this example it's obvious that more future rewards are important\
more you want to move in the direction of the end state.

The value of a state is the potential value that a given state can achieve\
